```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler

import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
```

```python
from sklearn.datasets import load_iris
data,target = load_iris(return_X_y=True,as_frame=True)
```

```python
print("Target dataframe head: ")
print(target.head())
```

```python
print("Input features dataframe tail: ")
print(data.tail())
```

```python
print("Splitting data for train and test - 80/20 split.")
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)
```

```python
print("Neural Network with 4 input neurons, 5 hidden layer neurons and 3 output neurons(1 for each class).")
model = Sequential([Dense(5,input_shape=(4,),activation='tanh'), Dense(3,activation='softmax')])
```

```python
print("Training model: ")
model.compile(optimizer = 'adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,validation_split=0.1,batch_size=2,epochs=150,shuffle=True,verbose=2)
```

```python
print("Model statistics: ")
print(model.summary())
```

```python
from sklearn.metrics import accuracy_score

print("Predicting on test set...")
y_predict = model.predict(X_test)

def printSoftmaxOutput(y_predict):
    classes = []
    for k in y_predict:
        cl = 0
        clval = 0
        for i in range(0, 3):
            if k[i] > clval:
                cl = i
                clval = k[i]
        classes.append(cl)
    return classes

y_pred_classes = printSoftmaxOutput(y_predict)
print("Predicted output classes: ")
print(y_pred_classes)
print("Corresponding real output classes: ")
print(y_test)
print("Accuracy of the model: ")
accuracy_score(y_test, y_pred_classes)
```
